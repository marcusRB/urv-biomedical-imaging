We can index the channels and show them separately

```python
import matplotlib.pyplot as plt

fig=plt.figure()
plt.subplot(1,3,1)
plt.imshow(image[:,:,0],cmap="jet")
plt.colorbar()
plt.subplot(1,3,2)
plt.imshow(image[:,:,1],cmap="gray")
plt.colorbar()
plt.subplot(1,3,3)
plt.imshow(image[:,:,2],cmap="gray")
plt.colorbar()
ax = fig.gca()
fig.set_size_inches((16,4))
```

Let us go back to our original image and check the concepts of sampling and quantization.

```python
from skimage import io

image = io.imread('./neighborhood6.jpg')
```

```python
image.shape
```

```python
im_sampled = image[::2,::2,:]
im_sampled.shape
```

```python
fig = plt.figure()
plt.subplot(1,2,1)
plt.imshow(image,interpolation='nearest')
plt.subplot(1,2,2)
plt.imshow(im_sampled,interpolation='nearest')
fig.set_size_inches((16,5))
```

```python
im_sampled = image[::8,::8,:]
im_sampled.shape
```

```python
fig = plt.figure()
plt.subplot(1,2,1)
plt.imshow(image,interpolation='nearest')
plt.subplot(1,2,2)
plt.imshow(im_sampled,interpolation='nearest')
fig.set_size_inches((16,5))
```

```python
im = image[0:100,0:30,:]
plt.imshow(im)
```

```python
import numpy as np
imgray = np.mean(image,2)
print(imgray.shape)


plt.imshow(imgray.astype('uint8'),cmap="winter")
plt.colorbar()
```
